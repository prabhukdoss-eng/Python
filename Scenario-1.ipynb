```python
def Amt_aftr_disc(order_amount):
    discount = 0.20  # 20% discount
    if order_amount > 100:
        discount_amount = order_amount * discount
        final_amount = order_amount - discount_amount
        print("Order amount :",order_amount , "\nAfter Discount :",final_amount)
    else:
        final_amount=order_amount
        print("Order amount :",order_amount , "\nAfter Discount :",final_amount)
        return final_amount
        
    
```

```python
Amt_aftr_disc(500)
```

```python
Amt_aftr_disc(50)
```

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return "Leap Year"
    else:
        return "Not Leap Year"
```

```python
is_leap_year(2025)
```

```python
is_leap_year(2028)
```

```python
def listofevenNos(nos):
    return [num for num in nos if num % 2 == 0]

listofvalues=[10,13,24,6,5,36,789]
even_numbers = listofevenNos(listofvalues)
print(even_numbers)  # Output: [2, 4, 6, 8, 10]
```
