```python
#Write logic to compute the sum of the digits of a given number
def sumOfDigits():
    num= input("Enter the passcode to open lock:") # Read the input number
    total = sum(int(d) for d in num if d.isdigit())
    print("Sum of digits:", total)
```

```python
sumOfDigits()
```
