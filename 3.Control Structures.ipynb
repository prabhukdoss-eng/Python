```python
age =int(input("Enter the age :"))
if(age <=18):
    print("Children")
elif(age>=19 and age<=50):
    print("adult")
elif(age>50):
    print("senor citizen")
else:
    print("Adult")
```

```python
lists = [10,20,30,40,50,60]
for age in lists:
    if (age <=18 ):
        print("Children")
    elif(age>=19 and age<=45):
        print("adult")
    elif(age>45 and  age<=55):
        print("middle age")
    else:
        print("Senior Citizen")
```
